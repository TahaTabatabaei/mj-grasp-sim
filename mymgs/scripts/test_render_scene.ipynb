### Scene rendering: robotiq_2f_85

make a new .npz file with this command;
`python -m mgs.cli.render_scene gripper=robotiq_2f_85 env=clutter_table`

find the output (a compressed 'render_*.npz' file) in the output folder (/mj-grasp-sim/mymgs)

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('/home/tahaos/Code/Projects/mj-grasp-sim/mymgs/MGS_OUTPUT_DIR/render_clutter_scene_33385ddc7cdb32c14101d3fe0a5cf566.npz', allow_pickle=True)

print(data.files)
```

```python
valid_grasps = data['valid_grasps']
print(valid_grasps.shape)
print(valid_grasps)
```

```python
scans = data['scans']
print(scans.shape)

for i in range(len(scans)):

    # plt.figure(figsize=(8, 8))  # Set width and height in inches
    plt.imshow(scans[i], cmap='gray')
    plt.title(f'Scan {i} RGBD')
    plt.show()
```

```python
masks = data['scan_masks']
print(masks.shape)
print(np.unique(masks))
```

```python
for i in range(10):

    # plt.figure(figsize=(8,8))
    plt.imshow(masks[i].astype(np.uint8))
    plt.title(f'Scan Mask {i}')
    plt.show()
```
