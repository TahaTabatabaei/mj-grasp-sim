### scan a gripper: robotiq_2f_85
make a new scan file with this commadn: \
`python -m mgs.cli.scan_gripper gripper=robotiq_2f_85`

find the output (a compressed .npz file) in the uotput folder (/mj-grasp-sim/mymgs)

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('/home/tahaos/Code/Projects/mj-grasp-sim/mymgs/MGS_OUTPUT_DIR/open_0e04b7eb557a73ba6f3cd52c7fa88606.npz')
```

```python
print(data.files)
```

```python

scans = data['scans']  # shape (10,480,480,4)
print(scans.shape)

# scans: image data
```

```python

for i in range(10):

    scan0 = scans[i]       # scan image
    # plt.figure(figsize=(8, 8))  # Set width and height in inches
    plt.imshow(scan0.astype(np.uint8))
    plt.title(f'Scan {i} RGBD')
    plt.show()

```

```python
scans_masks = data['scan_masks']  # shape (10,480,480,1)
print(scans_masks.shape)

# scan_masks: valid regions in image
```

```python
for i in range(10):

    # plt.figure(figsize=(8,8))
    plt.imshow(scans_masks[i].astype(np.uint8))
    plt.title(f'Scan Mask {i}')
    plt.show()
```

```python
segments = data['segments']
print(segments.shape)
print(segments)

# segments: pixel-wise labels for gripper parts (if defined)
```

```python
scan_extrinsics = data['scan_extrinsics']
print(scan_extrinsics.shape)
print(scan_extrinsics[0])

# scan_extrinsics: camera poses.
```

```python
scan_intrinsics = data['scan_intrinsics']
print(scan_intrinsics.shape)
print(scan_intrinsics[0])

# scan_intrinsics: camera calibration matrix.
```
